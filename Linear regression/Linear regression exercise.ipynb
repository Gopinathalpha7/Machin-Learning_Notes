# Linear regression exercise

## Steeps
* 1) Import libraries
* 2) Import the train data file and clean the data 
* 3) Separate features data
* 4) Separate Labels data
* 5) Create a linear Regression model
* 6) Fit the data into the model
* 7) Prediction result


```python
# 1) Import libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# 2) Import test data
train_data = pd.read_csv('canada_per_capita_income.csv')
```

```python
# 3) Separate features data
Future_data = train_data[['year']]
```

```python
# 4) Separate Labels data
Label_data = train_data['per capita income (US$)']
```

```python
# 5) Creat linear regretion model
lin_reg = linear_model.LinearRegression()
```

```python
# 6) Fit the data into model
lin_reg.fit(Future_data,Label_data)
```

```python
# 7) Prediction result (enter the area)
year = input('Enter year:')
lin_reg.predict([[year]])
```
