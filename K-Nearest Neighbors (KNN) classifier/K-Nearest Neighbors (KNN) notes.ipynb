# K-Nearest Neighbors (KNN) 
K-Nearest Neighbors (KNN) is a non-parametric machine learning algorithm that classifies or predicts a data point based on the majority class or average value of its K nearest neighbors in the feature space.

# K-Nearest Neighbors (KNN) Working principle

K-Nearest Neighbors (KNN) is a simple and intuitive machine learning algorithm used for classification and regression tasks. The primary idea behind KNN is to classify or predict a data point based on the majority class or average value of its K nearest neighbors in the feature space.

1. **Data Collection and Preparation:**
* Gather a labeled dataset with features (independent variables) and corresponding labels (class labels for classification or target values for regression).
* Preprocess the data, which may involve handling missing values, normalization, or feature scaling.

2. **Choosing the Value of K:**
* Select a value for K, which represents the number of nearest neighbors that will be considered for decision making.
* The choice of K is crucial, as a small K value may lead to noisy predictions, while a large K value may cause the boundaries between classes to blur.

3. **Calculating Distances:**
* Measure the distance between the data point to be classified (query point) and all other data points in the dataset.
* The most common distance metrics used are Euclidean distance for continuous data and Hamming distance for categorical data.

4. **Finding the K Nearest Neighbors:**
* Select the K data points with the shortest distances to the query point.
* These K data points are the nearest neighbors.

5. **Majority Voting (Classification) or Averaging (Regression):**
* For classification: Determine the class label by taking a vote among the K nearest neighbors. The class with the highest number of occurrences among the K neighbors is assigned to the query point.
* For regression: Calculate the average value of the target variable from the K nearest neighbors and assign that as the predicted value for the query point.

6. **Making Predictions:**
* For classification tasks, the predicted class label is assigned to the query point based on the majority voting result.
* For regression tasks, the predicted value is assigned based on the average value obtained from the K nearest neighbors.

7. **Model Evaluation:**
* Measure the performance of the KNN model using appropriate evaluation metrics such as accuracy (for classification) or mean squared error (for regression).
* Use cross-validation techniques to get a more reliable estimate of the model's performance.

8. **Prediction and Generalization:**
* Once the model is evaluated and deemed satisfactory, it can be used to make predictions on new, unseen data points.

It's important to note that KNN is a non-parametric and lazy learning algorithm. Non-parametric means that it doesn't assume any underlying distribution for the data, and lazy learning means that it doesn't learn a model during training; instead, it memorizes the entire training dataset and performs computations during the prediction phase. This aspect makes KNN easy to implement but potentially computationally expensive for large datasets.

# Over fitting and Under fitting

## What is over fitting ?

**Overfitting** - Perform good and high accuracy or very low training error in training data and perform bad and low accuracy or high error in test data

**Overfitting** is a common problem in machine learning and statistical modeling, where a model performs extremely well on the training data but fails to generalize accurately to new, unseen data. In other words, the model "memorizes" the training data instead of learning underlying patterns, leading to poor performance on real-world data.

In summary, overfitting is a phenomenon where a model becomes too specialized in the training data and loses its ability to generalize to new data, making it less useful for real-world applications. Preventing overfitting is essential for building robust and reliable machine learning models.

## What is Under fitting? 

**Underfitting** - A model which does not do well on both training and test data not at all learn the pattern is called an underfit model

(poor performance on training and test data (or) Losing both training and test data)

**Underfitting** is a phenomenon that occurs in machine learning and statistical modeling when a model is too simple or lacks the capacity to capture the underlying patterns in the data. It arises when the model is unable to sufficiently learn from the training data, leading to poor performance on both the training and unseen data (test/validation data).

In summary, underfitting is a situation in which a model is too simplistic to grasp the complexity of the data, resulting in poor performance on both the training set and unseen data. It is essential to find the right balance of model complexity and data availability to avoid underfitting.

## What is n-fold cross-validation method

n-fold cross-validation is a popular method used in machine learning and statistics to assess the performance of a predictive model and to mitigate potential overfitting issues. The "n" in n-fold represents the number of subsets into which the dataset is divided. Here's a step-by-step explanation of the n-fold cross-validation process for study purposes:

**Step 1**: Data Preparation
Start by gathering and preparing your dataset. Ensure that the data is clean, formatted correctly, and divided into input features (X) and target variable (y).

**Step 2**: Choosing the Value of n
Decide on the value of "n" for your cross-validation. Common choices are 5, 10, or more, depending on the size of your dataset and the computational resources available. Larger values of n provide more reliable results but require more computation.

**Step 3**: Dividing the Dataset
Divide your dataset into "n" subsets (folds) of approximately equal size. Each subset should have an equal representation of the target variable classes (if it's a classification problem) to ensure that each fold is representative of the entire dataset.

**Step 4**: Model Training and Evaluation
For each fold:
   a. Treat the current fold as the validation set.
   b. Combine the remaining "n-1" folds to form the training set.
   c. Train your machine learning model on the training set.
   d. Evaluate the model's performance on the validation set (the current fold).
   e. Record the evaluation metric(s) (e.g., accuracy, precision, recall, F1-score) to assess the model's performance.

**Step 5**: Average the Results
After repeating Step 4 for all "n" folds, you will have "n" evaluation metric values. Calculate the average of these values to get a more robust estimation of the model's performance.

**Step 6**: Model Selection and Hyperparameter Tuning (Optional)
If you are comparing multiple models or tuning hyperparameters, you can use n-fold cross-validation to assess their performance and choose the best-performing model or set of hyperparameters.

**Step 7**: Final Model Training
Once you have selected the best model or hyperparameters, train a final version of the model on the entire dataset (all "n" folds combined) to maximize the model's predictive power.

Advantages of n-fold Cross-Validation:
* It provides a more reliable estimate of the model's performance by leveraging the entire dataset for both training and evaluation.
* Helps in mitigating the risk of overfitting, as the model is tested on multiple independent datasets.
* Allows efficient use of data when the dataset is limited, as it repeatedly uses each data point for both training and validation.

Overall, n-fold cross-validation is a valuable technique for evaluating and selecting the best model in machine learning experiments, providing a more accurate assessment of a model's performance and generalization capabilities.

**(Note)** The number of folds depends on the data size. I usually stick with 4- or 5-fold (cv). Make sure to shuffle your data, such that your folds do not contain inherent bias.

```python
IQR = data['X'].quantile(0.75) - data['X'].quantile(0.25)
upper_threshold = data['X'].quantile(0.75) + (1.5*IQR)  # > q3 + 1.5iqr
lower_threshold = data['X'].quantile(0.25) - (1.5*IQR)  # < q1 - 1.5iqr

data['X'] = data['X'].clip(upper_threshold, lower_threshold) # eliminate out layers

upper_threshold, lower_threshold
```

```python
IQR = data['Y'].quantile(0.75) - data['Y'].quantile(0.25)
upper_threshold = data['Y'].quantile(0.75) + (1.5*IQR)  # > q3 + 1.5iqr
lower_threshold = data['Y'].quantile(0.25) - (1.5*IQR)  # < q1 - 1.5iqr

data['Y'] = data['Y'].clip(upper_threshold, lower_threshold) # eliminate out layers

upper_threshold, lower_threshold
```

```python
# Data Pre processing
# 6) Split the data
x = data[['X','Y']].values # Futures (X - axis)(2D Matrix)
y = data['class'].values  # Targe  (Y - axis)(1D Array)

# from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Data Pre processing
# 7) Scaling the data (Scaling mandatory for KNN)
    # 1) Import model
# from sklearn.preprocessing import StandardScaler
    
    # 2) Initialize
scaler = StandardScaler()
    
    # 3) Fit (Learning process)
scaler.fit(x_train)
    
    # 4) Transform
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# 8) Finding the best 'K' value by using cross-validation scores
# Create a list of K values to test (let's assume k varies from 1 to 30)
k_values = list(range(1, 31))

# Create an empty list to store the cross-validation scores for each K value
cv_scores = []

# Perform 5-fold cross-validation for each 'K' value
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k) #Initialising the model
    knn.fit(x_train,y_train) # Training the model
    scores = cross_val_score(knn, x_train_scaled, y_train, cv=5)
    cv_scores.append(scores.mean())
    print('K value:',k,  'train score :',knn.score(x_train_scaled,y_train),  'cv scores:',scores.mean(),)

# Find the best 'K' value with the highest cross-validation score
best_k = k_values[np.argmax(cv_scores)]
print('The highest cv score defines the best K value and the cv score is greater than the training score')
print('The Best K value is :', best_k)

# Plot the cross-validation scores for each 'K' value
plt.figure(figsize=(10, 6))
plt.plot(k_values, cv_scores, marker='o', linestyle='-', color='b')
plt.title('Cross-Validation Scores for Different K Values')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Cross-Validation Score')
plt.xticks(np.arange(1, 31, step=1))
plt.grid(True)

# Add a vertical line to indicate the best 'K' value
plt.axvline(x=best_k, color='r', linestyle='--', label=f'Best K value is = {best_k}')
plt.legend()

plt.show()
```
